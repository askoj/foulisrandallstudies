```python
# Aerts Example

from qogpsi import *

# States
N = state("N", label="North")
E = state("E", label="East")
S = state("S", label="South")
W = state("W", label="West")
NE = state("NE", label="North-East")
NW = state("NW", label="North-West")
SE = state("SE", label="South-East")
SW = state("SW", label="South-West")

# Random Variables
A = random_variable("A", states=[N,S])
A_prime = random_variable("A Prime", states=[E,W])
B = random_variable("B", states=[NE,SW])
B_prime = random_variable("B Prime", states=[NW,SE])

# Probability Space
p_s = probability_space(
    random_vars=[ A, A_prime, B, B_prime], 
    states=[ N, S, E, W, NE, NW, SE, SW])

# Contexts
p_s.event( "A|B", p={
    "N|NE" : 0.13,
    "N|SW" : 0.55,
    "S|NE" : 0.25,
    "S|SW" : 0.07,
})

p_s.event( "A|B Prime", p={
    "N|SE" : 0.47,
    "N|NW" : 0.12,
    "S|SE" : 0.06,
    "S|NW" : 0.35
})

p_s.event( "A Prime|B", p={
    "E|NE" : 0.13,
    "E|SW" : 0.38,
    "W|NE" : 0.42,
    "W|SW" : 0.07
})

p_s.event( "A Prime|B Prime", p={
    "E|SE" : 0.09,
    "E|NW" : 0.44,
    "W|SE" : 0.38,
    "W|NW" : 0.09
})

# Distribution
distribution = p_s.distribution(
    x=[A, A_prime],
    y=[B, B_prime]
)

FR = foulis_randall([
    [A, A_prime],
    [B, B_prime]
])

distribution.graph(highlight=FR["hyperedges"], scale=1)
```

```python
NO_graph = non_orthogonality(FR=FR)
NO_graph.graph()
```

```python
NO_graph.is_classical(p_s)
```
